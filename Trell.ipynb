```python
import pandas as pd
import numpy as np
```

```python
#searchTrails Analysis
searchTrails = pd.read_csv('SearchAnalysis.csv')
searchTrails.head()
```

```python
searchTrails = searchTrails.fillna(0)
searchTrails['Likes'] = searchTrails['Likes'].astype(int)
searchTrails['Comments'] = searchTrails['Comments'].astype(int)
searchTrails.head()
```

```python
searchTrails.describe()
```

```python
st_freq = searchTrails.groupby(['InterestType']).size()
st_freq
```

```python
st_freq.plot.bar()
```

```python
st_avg = searchTrails.groupby(['InterestType']).mean()
st_avg
```

```python
st_avg.plot.bar(y=['Views'])
```

```python
st_avg.plot.bar(y=['Likes'])
```

```python
st_avg.plot.bar(y=['Posts'])
```

```python
st_sum = searchTrails.groupby(['InterestType']).sum()
st_sum
```

```python
st_sum['Likes/View'] = st_sum['Likes']/st_sum['Views']
st_sum
```

```python
st_sum.plot.bar(y=['Likes/View'])
```
